```python
num_cols = [
    'course_duration_hours',
    'rating',
    'enrollment_numbers',
    'course_price',
    'feedback_score'
]
```

```python
# Before outlier handling
for col in num_cols:
    sns.boxplot(x=data[col])
    plt.title(f"Before Outlier Handling - {col}")
    plt.show()
```

```python
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    data = data[
        (data[col] >= Q1 - 1.5 * IQR) &
        (data[col] <= Q3 + 1.5 * IQR)
    ]
```

```python
# After outlier handling
for col in num_cols:
    sns.boxplot(x=data[col])
    plt.title(f"After Outlier Handling - {col}")
    plt.show()
```

```python
data.reset_index(drop=True, inplace=True)
data['row_pos'] = np.arange(len(data))

print("After outlier handling:", data.shape)
```

#### Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
ratings = data[['user_id', 'course_id', 'rating']].copy()

from sklearn.model_selection import train_test_split
train, test_df = train_test_split(
    ratings, test_size=0.2, random_state=42
)

```

```python
global_mean = train['rating'].mean()
```

#### MODEL 1  POPULARITY BASED

```python
test_df['popularity_pred'] = global_mean

popularity_rmse = np.sqrt(
    mean_squared_error(test_df['rating'], test_df['popularity_pred'])
)
```

```python
popularity_rmse
```

#### MODEL 2   CONTENT BASED


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
global_mean = train['rating'].mean()
```

```python
def content_predict(row):
    ratings = train[
        train['course_id'] == row['course_id']
    ]['rating']
    
    return ratings.mean() if not ratings.empty else global_mean
```

```python
test_df['content_pred'] = test_df.apply(
    content_predict, axis=1
)
```

```python
content_rmse = np.sqrt(
    mean_squared_error(
        test_df['rating'],
        test_df['content_pred']
    )
)

print("Content-Based RMSE:", content_rmse)
```

#### MODEL 3  Collaborative Filtering

```python
user_bias = train.groupby('user_id')['rating'].mean() - global_mean
item_bias = train.groupby('course_id')['rating'].mean() - global_mean
```

```python
def cf_predict(row):
    bu = user_bias.get(row.user_id, 0)
    bi = item_bias.get(row.course_id, 0)
    return global_mean + bu + bi

test_df['cf_pred'] = test_df.apply(cf_predict, axis=1)

cf_rmse = np.sqrt(
    mean_squared_error(test_df['rating'], test_df['cf_pred'])
)
```

```python
cf_rmse
```

#### MODEL 4  HYBIRD

```python
def hybrid_predict(row):
    return 0.5 * cf_predict(row) + 0.5 * content_predict(row)

test_df['hybrid_pred'] = test_df.apply(hybrid_predict, axis=1)

hybrid_rmse = np.sqrt(
    mean_squared_error(test_df['rating'], test_df['hybrid_pred'])
)
```

```python
hybrid_rmse
```

#### Model Comparision

```python
results = pd.DataFrame({
    'Model': [
        'Popularity',
        'Content-Based',
        'Collaborative Filtering',
        'Hybrid'
    ],
    'RMSE': [
        popularity_rmse,
        content_rmse,
        cf_rmse,
        hybrid_rmse
    ]
}).sort_values(by='RMSE')

print("\nMODEL COMPARISON")
print(results)
```

#### Final result

```python
def get_top_n(user_id, n=5):
    seen = train[train.user_id == user_id]['course_id']
    candidates = item_mean.index.difference(seen)

    scores = []
    for course in candidates:
        row = pd.Series({'user_id': user_id, 'course_id': course})
        scores.append((course, hybrid_predict(row)))

    return sorted(scores, key=lambda x: x[1], reverse=True)[:n]
```

```python
def get_top_n(user_id, n=5):
    # SAFETY: recompute item_mean if it doesn't exist
    item_mean_local = train.groupby('course_id')['rating'].mean()

    seen = train[train.user_id == user_id]['course_id']
    candidates = item_mean_local.index.difference(seen)

    scores = []
    for course in candidates:
        row = pd.Series({'user_id': user_id, 'course_id': course})
        scores.append((course, hybrid_predict(row)))

    return sorted(scores, key=lambda x: x[1], reverse=True)[:n]
```

```python
course_lookup = (
    data[['course_id', 'course_name', 'instructor', 'rating']]
    .drop_duplicates(subset='course_id')
)
```

```python
def get_top_n_readable(user_id, n=5):
    recs = get_top_n(user_id, n)

    rec_df = pd.DataFrame(
        recs, columns=['course_id', 'recommendation_score']
    )

    result = rec_df.merge(
        course_lookup,
        on='course_id',
        how='left'
    )

    return result.sort_values(
        by='recommendation_score',
        ascending=False
    )
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Ensure data is loaded (assuming 'df' is your main dataframe)
# If your dataframe has a different name, change 'df' to that name.
train_data = df.copy()

# 2. Calculate Hybrid Model Components
# These are essential for the 'Personalized' part of your recommender
global_mean = train_data['rating'].mean()
user_bias = (train_data.groupby('user_id')['rating'].mean() - global_mean).to_dict()
item_bias = (train_data.groupby('course_id')['rating'].mean() - global_mean).to_dict()
user_history = train_data.groupby('user_id')['course_id'].apply(list).to_dict()

# 3. Handle Content-Based Similarity (Memory-Safe Version)
# We calculate similarity for UNIQUE courses only to avoid the 74GB MemoryError
unique_courses = train_data[['course_id', 'course_name', 'instructor', 'rating']].drop_duplicates('course_id').reset_index(drop=True)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(unique_courses['course_name'].fillna(''))
similarity = cosine_similarity(tfidf_matrix)

# 4. Save everything into 3 specific Pickle files
# Use 'wb' (Write Binary) to avoid encoding issues
try:
    # File 1: The course database for the UI
    pickle.dump(unique_courses, open('courses.pkl', 'wb'))
    
    # File 2: The similarity scores for "Similar Courses" feature
    pickle.dump(similarity, open('similarity.pkl', 'wb'))
    
    # File 3: The mathematical biases for "Personalized" feature
    model_logic = {
        'global_mean': global_mean,
        'user_bias': user_bias,
        'item_bias': item_bias,
        'user_history': user_history
    }
    pickle.dump(model_logic, open('model_logic.pkl', 'wb'))
    
    print(" All 3 files generated successfully!")
    print("1. courses.pkl")
    print("2. similarity.pkl")
    print("3. model_logic.pkl")
    
except Exception as e:
    print(f" Error saving files: {e}")
```
